``用户自定义类``
Grid 类：并行显示多张图 
Overlap 类：结合不同类型图表叠加画在同张图上 
Page 类：同一网页按顺序展示多图 
Timeline 类：提供时间线轮播多张图

```python
from pyecharts import Bar,Line, Overlap

CLOTHES = ["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"]
clothes_v1 = [5, 20, 36, 10, 75, 90]
clothes_v2 = [10, 25, 8, 60, 20, 80]

bar = Bar("Line - Bar")
bar.add("bar", CLOTHES, clothes_v1, is_more_utils=True,is_xaxis_boundarygap=True)
line = Line()
line.add("line", CLOTHES, clothes_v2)

overLap = Overlap()
overLap.add(bar)
overLap.add(line)
```

Graph关系图的展示

```python
from pyecharts import Graph

nodes = [{"name": "结点1", "symbolSize": 10, "value":1},
         {"name": "结点2", "symbolSize": 20, "value":2},
         {"name": "结点3", "symbolSize": 30, "value":3},
         {"name": "结点4", "symbolSize": 40, "value":4},
         {"name": "结点5", "symbolSize": 50, "value":5},
         {"name": "结点6", "symbolSize": 40, "value":6},
         {"name": "结点7", "symbolSize": 30, "value":7},
         {"name": "结点8", "symbolSize": 20, "value":8}]
links = []
for i in nodes:
    for j in nodes:
        links.append({"source": i.get('name'), "target": j.get('name')})
graph = Graph("关系图-力引导布局示例")
graph.add("", nodes, links, repulsion=8, 
          is_roam= True, graph_edge_length=100,
          graph_gravity=1,graph_repulsion=1000,
          is_rotatelabel=True,categories=[1,2,3,4,5,6],
          graph_layout="force",is_visualmap=True)
```
